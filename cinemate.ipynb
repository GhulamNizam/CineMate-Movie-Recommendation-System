```python
# some features containn python literal so we will extract only name's value in each dictionaries keys
```

```python
import ast
```

```python
def convert(text):
    L = []
    for i in ast.literal_eval(text):
        L.append(i['name'])
    return L
```

```python
movies['genres'] = movies['genres'].apply(convert)
movies['keywords'] = movies['keywords'].apply(convert)
```

```python
#we will extract only top 3 actors' name in the movie from cast column
```

```python
def TopActorsName(text):
    counter =0
    L = []
    for i in ast.literal_eval(text):
        if counter < 3:
            L.append(i['name'])
        else:
            break
        counter = counter+1
    return L
```

```python
movies['cast'] = movies['cast'].apply(TopActorsName)
movies.head()
```

```python
#now will fetch dirctor from the crew
def FetchDirector(text):
#     counter =0
    L = []
    for i in ast.literal_eval(text):
        if i['job'] == 'Director':
            L.append(i['name'])
#         else:
#             break
#         counter = counter+1
    return L
```

```python
movies['crew'] = movies['crew'].apply(FetchDirector)
```

```python
# id is not rqd
movies.pop('id')
```

```python
movies.head()
```

```python
#overview is a string;converting it to list
movies['overview']=movies['overview'].apply(lambda x:x.split())
```

```python
# In keywords, cast, crew & genre theres a space in names 
# when we will stemm the words it will create two diff tages for one name only e.g; Sam Worthington->Sam,Worthington
# so to avoid this we will remove the space and make it one entity
```

```python
def RemoveSpace(text):
    L = []
    for i in text:
        L.append(i.replace(" ",""))
    return L
    
```

```python
movies['genres']=movies['genres'].apply(RemoveSpace)
movies['crew']=movies['crew'].apply(RemoveSpace)
movies['cast']=movies['cast'].apply(RemoveSpace)
movies['keywords']=movies['keywords'].apply(RemoveSpace)
```

```python
movies.head()
```

```python
# making one feature including characterstics of movie
movies['tags'] = movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew']
```

```python
new = movies.drop(columns=['overview','genres','keywords','cast','crew'])
#new.head()
```

```python
# joining them to make a para
new['tags'] = new['tags'].apply(lambda x: " ".join(x))
new.head()
```

```python
#reordering
# Define the desired order of features (columns)
desired_order = ['movie_id', 'title', 'tags']

# Select the columns of the DataFrame in the desired order
new = new[desired_order]
```

```python
new.head()
```

### model building

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
CV = CountVectorizer(max_features=5000,stop_words='english')
```

```python
vector = CV.fit_transform(new['tags']).toarray()
```

```python
vector.shape
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vector)
```

```python
similarity
```

```python
new[new['title'] == 'The Lego Movie'].index[0]
# it will give row number of given movie e.g, 'The Lego Movie':744
```

```python
# similarity increases when distance decreases
# to predict similar movies we will find top 5 nearest movies of the given movies descending order of similarity
def predict(movie):
    index = new[new['title'] == movie].index[0]
    distances = sorted(list(enumerate(similarity[index])),reverse=True,key = lambda x: x[1])
    
    for i in distances[1:6]:
        print(new.iloc[i[0]].title)
```
